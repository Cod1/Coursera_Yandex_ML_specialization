# Central limit theorem

```python
import numpy as np
import scipy.stats as sts
import math
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Rayleigh distribution (https://en.wikipedia.org/wiki/Rayleigh_distribution)

## Generate sample size of 1000

```python
rayleigh = sts.rayleigh(loc=0, scale=1)
sample = rayleigh.rvs(1000)
```

# Histogram and PDF

```python
x = np.linspace(0, 5, 1000)
```

```python
#sample histogram
plt.hist(sample, bins=17, normed=True)

#theoretical pdf
pdf = rayleigh.pdf(x)
plt.plot(x, pdf, label='pdf', alpha=0.5, color='red')

plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

# Central limit theorem in practice

## Generate samples size of n = [3, 10, 20, 35, 50]

```python
sigma = 1
```

```python
th_mean = math.sqrt(math.pi / 2) * sigma #theoretical mean
th_var = (2 - math.pi / 2) * sigma * sigma #theoretical variance
```

```python
n_array = [3, 10, 20, 35, 50] #size of generating sample
```

```python
for n in n_array:
    sample5 = []
    for i in range(0, 1001):
        tmp_sample = rayleigh.rvs(n)
        cur_mean = np.mean(tmp_sample)
        sample5.append(cur_mean)
    
    #sample histogram
    plt.hist(sample5, bins=17, normed=True)
    
    #normal distribution
    norm = sts.norm(th_mean, math.sqrt(th_var/n))
    pdf = norm.pdf(x)

    #normal distribution pdf plot
    pdf = norm.pdf(x)
    plt.plot(x, pdf, label='pdf', alpha=0.5, color='red')
    
    plt.title('n = ' + str(n))
    plt.legend()
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.show()
```

# Summary

### The accuracy of the approximation of the distribution of sample normal means increases with n by CLT theorem
